```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
import math
```

```python

a_x = np.random.rand(50,1)*1
a_y = [1-x for x in a_x] + np.random.rand(50,1)*1
b_x = np.random.rand(50,1)*2 +1.5
b_y = np.random.rand(50,1)*2 +1.5

fig, ax = plt.subplots()
ax.scatter(a_x,a_y,c = 'r',marker = 'o')
ax.scatter(b_x,b_y,c = 'b',marker = 'o')
plt.show()
```

```python
a = np.column_stack([a_x,a_y])
b = np.column_stack([b_x,b_y])

features = np.concatenate((a,b))
labels = np.concatenate((np.zeros(50)-1,np.zeros(50)+1))

clf = svm.LinearSVC()
clf.fit(features,labels)

x = np.linspace(-1, 4, 200)
y = np.linspace(-1, 4, 200)
# 生成网格
X,Y = np.meshgrid(x, y)
res = clf.predict(np.c_[X.ravel(), Y.ravel()]).reshape(X.shape)

fig, ax = plt.subplots()
# 等高线
ax.contourf(X, Y, res,alpha=0.8)
ax.scatter(a_x,a_y,c = 'r',marker = 'o')
ax.scatter(b_x,b_y,c = 'b',marker = 'o')
```

```python
a = np.random.rand(50) * math.pi/2 
a_x = np.sin(a) + 0.5*np.random.rand(50)
a_y = np.cos(a) + 0.5*np.random.rand(50)

b = np.random.rand(50) * math.pi/2 
b_x = np.sin(b)*2 + np.random.rand(50)
b_y = np.cos(b)*2 + np.random.rand(50)

c = np.random.rand(50) * math.pi/2 
c_x = np.sin(b)*3 + 0.5*np.random.rand(50)
c_y = np.cos(b)*3 + 0.5*np.random.rand(50)

fig, ax = plt.subplots()
ax.scatter(a_x,a_y,c = 'r')
ax.scatter(b_x,b_y,c = 'b')
ax.scatter(c_x,c_y,c = 'g')
plt.show()
```

```python
a = np.column_stack([a_x,a_y])
b = np.column_stack([b_x,b_y])
c = np.column_stack([c_x,c_y])

features = np.concatenate((a,b,c))
labels = np.concatenate((np.zeros(50),np.zeros(50)+1,np.zeros(50)+2))

clf = svm.SVC(kernel='linear', C=1)
clf.fit(features,labels)

clf2 = svm.SVC(kernel='rbf', gamma=0.7, C=1)
clf2.fit(features,labels)

x = np.linspace(0, 4, 200)
y = np.linspace(0, 4, 200)
# 生成网格
X,Y = np.meshgrid(x, y)
res1 = clf.predict(np.c_[X.ravel(), Y.ravel()]).reshape(X.shape)
res2 = clf2.predict(np.c_[X.ravel(), Y.ravel()]).reshape(X.shape)

figs, axs = plt.subplots(2)
# plt.subplots_adjust(wspace=0.4, hspace=0.4)
# 等高线
for ax,res in zip(axs,[res1,res2]):
    ax.contourf(X, Y, res,alpha=0.8)
    ax.scatter(a_x,a_y,c = 'r')
    ax.scatter(b_x,b_y,c = 'b')
    ax.scatter(c_x,c_y,c = 'g')
```
